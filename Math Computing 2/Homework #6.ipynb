2. Write a function to draw three heights for a acute triangle via turtle. The input of the function are the coordinates of three vertexes. Hint: https://en.wikipedia.org/wiki/Distance_from_a_point_to_a_line.

For any triangle with sides a, b, c and semiperimeter s = (a + b + c) / 2, the altitude from side a is given by

```python
import turtle
win=turtle.Screen()
win.setworldcoordinates(-10,-10,10,10)
t=turtle.Turtle()
t.hideturtle()
t.speed(20000)

def triangleheights(v1x,v1y,v2x,v2y,v3x,v3y):
    
    #compute side lengths
    a=((v1x-v2x)**2+(v1y-v2y)**2)**.5
    b=((v3x-v2x)**2+(v3y-v2y)**2)**.5
    c=((v1x-v3x)**2+(v1y-v3y)**2)**.5
    
    #draw main acute triangle for the given vertices
    t.up()
    t.goto((v1x,v1y))
    t.down()
    t.goto((v2x,v2y))
    t.goto((v3x,v3y))
    t.goto((v1x,v1y))
    
    
    t.setheading(t.towards(v2x, v2y)) #point to vertice B
    t.forward(a/2) #travel turtle to midpoint
    t.goto((v3x,v3y)) #cut across to vertice C
    t.up()
    t.goto((v2x,v2y))
    t.down()
    t.setheading(t.towards(v3x, v3y))
    t.forward(b/2)
    t.goto((v1x,v1y))
    t.setheading(t.towards(v3x, v3y))
    t.forward(c/2)
    t.goto((v2x,v2y))
    
    
triangleheights(10,1,7,7,-10,1)
win.exitonclick()    
```
