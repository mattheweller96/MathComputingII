```python
import math
#approx mod function for dx, input the function and an x
def dx(f, x):
    return abs(0-f(x))
def function(x):
    return ((x**4) - (3*x**3) + (2*math.sin(x)))
def derivative(x):
    return ((4*x**3) - (9*x**2) + (2*math.cos(x)))
def Newton_method(start,function,derivative,tolerance):
    change = dx(function,start)
    while change > tolerance:
        start = start - function(start)/derivative(start)
        change = dx(function,start)
    print('Root:',start)
    print('function evaluated at root is:',function(start))

#function will have three roots so we use three guesses and then we can compare
#to find the minimal root
starts = [0,.5,1]
for start in starts:
    Newton_method(start, function,derivative,0.0000005)
#our minimum root value is in the output and the evaluation in the function is also
#in the output prints.
```
