```
1. Draw a (time series) line plot for the daily number of passengers in the JFK airport during

```

````python
airport=pd.read_csv("C:/Users/Matthew/Documents/Airport_Wait_Time.csv")
airport["Date"]=pd.to_datetime(airport['Date'])

passengersall = DataFrame(airport,
              columns=['Airport','Date','Number of Passengers'])
alljfk=passengersall.loc[passengersall.Airport=='JFK']
alljfk=alljfk.set_index(['Date'])
jfk=alljfk.loc['2017-12-14':'2018-1-7']
jfkdates=jfk.groupby(by='Date')
gg=jfkdates.sum()
````

```
3. For each airport in the dataset, find the average daily number of passengers, and the average daily number of booths. Make a scatter plot between the two variables where each dot represents for an airport. What do you find from the plot?
```

````python
airport=pd.read_csv("C:/Users/Matthew/Documents/Airport_Wait_Time.csv")
````

````python
def b(x):
    return int(sum(x['Number of Booths']/x['Date'].count()))

avg_booths = airport.groupby(['Airport']).apply(b).reset_index() 
avg_booths.columns=['Airport','AvgBooths']
avg_booths = avg_booths.set_index(['Airport'])


def g(x):
    return int(sum(x['Number of Passengers']/x['Date'].count()))
avg_pass = airport.groupby(['Airport']).apply(g).reset_index()
avg_pass.columns=['Airport','AvgPass']
avg_pass = avg_pass.set_index(['Airport'])
avg_daily = pd.concat([avg_booths,avg_pass],axis=1)



groups = avg_daily.groupby('Airport')
fig=plt.figure(figsize=(18,12.5)); ax=fig.add_subplot(1,1,1)
ax.margins(0.15)

for name, group in groups:
    ax.plot(group.AvgBooths, group.AvgPass, marker='o', linestyle='', ms=10, label=name)
    
    
ax.legend(loc='best')
ax.set_xlabel('Avg Daily Booths')
ax.set_ylabel('Avg Daily Passengers')
ax.set_title('Daily Number of Passengers per Daily Number of Booths at Airports')
plt.show()
````

## As we can see from the above graph, it appears the average ratio between the number of booths and the number of passengers is a constant with small deviation.

```
Use regular expression to process the data so that the camel-case words can be turned back to the normal convention with spaces inserted. Below are a few requirements:
```

```
    a) In general case, add a space between two words, e.g. "DouglasCounty"  ---> "Douglas County".

    b) If two words are linked by a dash, then keep the dash and do not add the space. 
    E.g. "Exeter-MilliganPublicSchools" ---> "Exeter-Milligan Public Schools".

    c) If two words are linked by a dot, then append the dot with the previous word and add a space. 
    E.g. "St.PaulPublicSchool" ---> "St. Paul Public School".

    d) If there is a number, then separate the number as a word. 
    E.g. "WallacePublicSchoolDistrict65R" ---> "Wallace Public School District 65 R".
```

````python
school_file = open("C:/Users/Matthew/Documents/NE_School_Districts.csv","r")

newschool_file = open("C:/Users/Matthew/Documents/NE_School_Districts.txt", "w")

for line in school_file:
    
  
    line = re.sub("([a-z])([A-Z])","\g<1> \g<2>",line)  #adding spaces for capital letters
    line = re.sub("([^-])([A-Z][a-z-]+)", r"\1 \2", line)  #adding spaces after dashes
    line = re.sub("(Mc|Mac)(\s)", r"\1", line)  #keeping Mc
    line = re.sub(r"([0-9]+(\.[0-9]+)?)",r" \1 ", line) #adding spaces after a digit with a nondigit next to it
    
    print(line)
    
    newschool_file.write(line)
    newschool_file.write("\n")
    
newschool_file.close()
school_file.close()
````
